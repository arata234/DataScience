10-100の整数データ(size=100)からSeriesを作成

```python
arr_int2 = np.random.randint(10, 100, 100)
sr_quantity = pd.Series(arr_int2, name="Quantity")
```

0-10の小数データ(100, 2)からDFを作成    

```python
arr_fl = np.random.rand(100, 2) * 10
# print(arr_fl)
```

```python
df = pd.DataFrame(arr_fl)
df
```

```python
df_wh = pd.DataFrame(arr_fl, columns=["Width", "Height"])
df_wh
```

40-60の範囲の一様分布の少数データ(100, 1)からDFを作成

```python
arr_uni = np.random.uniform(40, 60, 100)
df_quality = pd.DataFrame(arr_uni, columns=["Quality"])
df_quality
```

NormDist(mean=0, std=1)に従う小数データ(100, 1)からDFを作成

```python
arr_norm = np.random.normal(loc=0, scale=1, size=(100, 1))
df_difference = pd.DataFrame(arr_norm, columns=["Difference"])
df_difference
```

['red', 'green', 'blue'](100, 1), ['circle', 'square', 'triangle'](100, 1), ndarryを結合しDataFrameを作成

```python
colors = ['red', 'green', 'blue']
shapes = ['circle', 'square', 'triangle']

arr_colors = np.random.choice(colors, size=(100, 1))
arr_shapes = np.random.choice(shapes, size=(100, 1))

arr_cs = np.concatenate([arr_colors, arr_shapes], axis=1)
arr_cs_hstack = np.hstack([arr_colors, arr_shapes])
arr_cs_vstack = np.vstack([arr_colors, arr_shapes])
```

```python
df_cs = pd.DataFrame(arr_cs, columns=["Colors", "Shape"])
df_cs
```

datetime 1977/7/5から1日刻み(100, 1)からDatetimeIndex, Seriesを作成

```python
dates = pd.date_range('1997/7/5', periods=100, freq="D")
sr_date = pd.Series(dates, name="Date")
sr_date
```

Series, DataFrameを結合

```python
df = pd.concat([sr_date, sr_price, sr_quantity, df_wh, df_quality, df_difference], axis=1)
```
