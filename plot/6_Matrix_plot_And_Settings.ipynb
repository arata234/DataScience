## 2. Numpyのwhere(), select()を用いたレコード抽出と置換

* np.where(condition, True-value, False-value)
* np.select(conditions, values)  
---

np.where(condition, True-value, False-value)

```python
df[df['petal_width'] > 1.3]
```

```python
np.where(df['petal_width'] > 1.3, 'wide', 'narrow')
```

np.select(conditions, values)

```python
conditions = [df['petal_width'] < 1, 
              (1 <= df['petal_width']) & (df['petal_width'] <= 2),
              2 < df['petal_width']]

values = ['narrow', 'medium', 'wide']
```

```python
np.select(condlist=conditions, choicelist=values)
```

```python
df['pw_cat'] = np.select(conditions, values)
```

```python
# 後のプロットで使うため、ランダムな分類を設定しておく
np.random.seed(17)
arr = ['red', 'green', 'blue']
df['color'] = np.random.choice(arr, len(df))
df.head()
```

Seabornテーマの設定

```python
sns.set_theme(context='talk', style='darkgrid', font='MS GOTHIC')
```

ヒートマップ

```python
df_corr = df.corr()
sns.heatmap(data=df_corr, robust=True, annot=True, cmap='Spectral')
plt.show()
```

クラスターマップ  
※ 属性が似た

```python
sns.clustermap(data=df_corr, robust=True, annot=True, 
               vmin=-1,cmap='Spectral')
plt.show()
```

ペアグリッド  
※ グリッド形式で2値の組み合わせ毎の関係を表す

```python
g = sns.PairGrid(data=df, hue='species', vars=df.select_dtypes('float').columns,
                 height=4)
g.map_diag(sns.histplot, multiple='fill',stat='probability')
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot, alpha=0.6)
plt.legend()
plt.show()
```

```python
length = ['petal_length', 'sepal_length']
width = ['petal_width', 'sepal_width']
```

```python
g = sns.PairGrid(data=df, hue='species', x_vars=length, 
                 y_vars=width, height=4, aspect=2)
g.map(sns.scatterplot, alpha=0.6)
g.add_legend()
plt.show()
```

Facetグリッド  
※ row と col にクラスで分類しそれぞれにプロット

```python
df.head()
```

```python
g = sns.FacetGrid(data=df, row='pw_cat', col='color', height=4)
g.map_dataframe(sns.histplot, x='sepal_length', hue='species', multiple='stack')
g.add_legend()
plt.show()
```

## 4. パイチャートと画像の保存

```python
data = df['species'].value_counts()
labels = df['species'].unique()
labels
```

画像の保存

```python
# startangle: 開始位置
# conterclock: 
#
plt.figure(figsize=(10, 8), facecolor='#444')
plt.pie(data, labels=labels, autopct='%.2f%%',
        textprops={'color': '#fff', 'size':20}, startangle=90,
        counterclock=False, explode=[0, 0, 0.1], pctdistance=0.5)
plt.title('PIE', size=40, color='#fff')
plt.savefig('./data/pie.png', dpi=100)
plt.show()
```

## 5. グラフの表示調整

```python
setosa = df[df['species'] == 'setosa']
versicolor = df[df['species'] == 'versicolor']
virginica = df[df['species'] == 'virginica']
```

```python
plt.figure(figsize=(8, 4), facecolor='#abc', dpi=150)
# sns.scatterplot(data=df, x='petal_width', y='petal_length', hue='species', alpha=0.6)
sns.regplot(data=setosa, x='petal_width', y='petal_length', truncate=True, label="setosa", scatter_kws={'alpha': 0.3})
sns.regplot(data=versicolor, x='petal_width', y='petal_length', truncate=True, label="versicolor", scatter_kws={'alpha': 0.3})
sns.regplot(data=virginica, x='petal_width', y='petal_length', truncate=True, label="virginica", scatter_kws={'alpha': 0.3})
plt.title('PETAL', fontsize=30, color='#fff', pad=20)
plt.xlabel('WIDTH', fontsize=20)
plt.ylabel('HEIGHT', fontsize=20)

plt.xticks(np.arange(0, 3, 0.5))
plt.yticks([2, 4, 6], ['short', 'medium', 'tail'])
plt.axhline(df['petal_length'].mean(), color='red', linewidth=1, linestyle='--')
plt.axvline(df['petal_width'].mean(), color='red', linewidth=1, linestyle='--')
plt.legend(loc='lower right', title='species', fontsize=12, title_fontsize=14, bbox_to_anchor=(0.9, 1,))
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8), facecolor='#abc', dpi=80, tight_layout=True)

sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', ax=axes[0,0])
sns.scatterplot(data=df[df['color']=='red'], x='petal_length', y='petal_width', hue='species', ax=axes[0,1], palette='Reds')
sns.scatterplot(data=df[df['color']=='green'], x='petal_length', y='petal_width', hue='species', ax=axes[1,0], palette='Greens')
sns.scatterplot(data=df[df['color']=='blue'], x='petal_length', y='petal_width', hue='species', ax=axes[1,1], palette='Blues')
```
