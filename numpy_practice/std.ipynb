```python
import numpy as np
```

```python
np.random.randn(1000).std()
```

```python
np.random.seed(42)
import matplotlib.pyplot as plt
plt.hist(np.random.randn(1000), bins=30)
plt.show()
```

np.random.normal(mean, std, size):  正規分布に近づく数列

```python
np.random.normal(50, 10, 10)
```

```python
np.random.normal(50, 10, (2, 5))
```

```python
fig = plt.figure(figsize=(6, 4), dpi=300)
ax = fig.add_subplot()

ax.hist(np.random.normal(100, 4, 1000), bins=30)
ax.set_xlim(80, 120)
plt.show()
```
